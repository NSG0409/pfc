```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

from sklearn.preprocessing import StandardScaler
```

```python
import numpy as np
```

```python
def prepare_data(series, n_lags, n_ahead):
    X, y = [], []
    for i in range(len(series) - n_lags - n_ahead + 1):
        X.append(series[i:i + n_lags])
        y.append(series[i + n_lags:i + n_lags + n_ahead])
    return np.array(X), np.array(y)
```

```python
def evaluate_models(X_train, y_train, X_test, y_test, models):
    results = {}
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        results[name] = {'MSE': mse, 'MAE': mae, 'Predictions': y_pred}
    results
```

```python
models = {
    'LinearRegression': LinearRegression(),
    'RandomForest': RandomForestRegressor(),
    'SVR': SVR()
}

metrics = ['MSE', 'MAE']
```

```python
def plot_predictions(y_test, predictions, model_name):
    plt.figure(figsize=(10,5))
    plt.plot(y_test, label='True Values', color='blue')
    plt.plot(predictions, label=f'{model_name} Predictions', color='red')
    plt.title(f'Predictions vs True values for {model_name}')
    plt.legend()
    plt.show()
    
```

```python
def prediction_pipeline(data, target_column, n_lags, n_ahead):
    # 1 - target a definir
    series = data[target_column].values
    
    # 2 - preparar os dados
    X, y = prepare_data(series, n_lags, n_ahead)
    
    # 3 - Dividir em treino e teste
    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size=0.2,
        shuffle=False,
        random_state=42
    )
    
    # 4 - Testar os modelos
    results = evaluate_models(X_train, y_train, X_test, y_test, models)
    
    # 5 - Mostrar e plotar os resultados
    for model_name, result in results.items():
        print(f'{model_name} - MSE: {result["MSE"]:.4f}, MAE: {result["MAE"]:.4f}')
        plot_predictions(y_test, result['predictions'], model_name)
```

```python
prediction_pipeline(df_2001, target_column='T2M', n_lags=1000, n_ahead=1)
```
